### Data Preprocessing

#### Convert Attrition from ('Yes', 'No') to (1,0) 

```python
def convertAttrition(x):
    if x == 'Yes':
        return 1
    else:
        return 0
```

```python
df['Attrition'] = df['Attrition'].apply(convertAttrition)
df.head()
```

### Convert the rest of the categorical values into numeric using dummy variables and store the results in a new dataframe called 'newdf'

```python
newdf = pd.get_dummies(df, drop_first = True).astype(int)
newdf.head()
```

#### Check the shape of our new dataset

```python
newdf.shape
```

#### Print unique values in our new dataframe

```python
unique_values = ['Age', 'Attrition', 'DailyRate', 'DistanceFromHome', 'Education',
       'EnvironmentSatisfaction', 'HourlyRate', 'JobInvolvement', 'JobLevel',
       'JobSatisfaction', 'MonthlyIncome', 'MonthlyRate', 'NumCompaniesWorked',
       'PercentSalaryHike', 'PerformanceRating', 'RelationshipSatisfaction',
       'StockOptionLevel', 'TotalWorkingYears', 'TrainingTimesLastYear',
       'WorkLifeBalance', 'YearsAtCompany', 'YearsInCurrentRole',
       'YearsSinceLastPromotion', 'YearsWithCurrManager',
       'BusinessTravel_Travel_Frequently', 'BusinessTravel_Travel_Rarely',
       'Department_Research & Development', 'Department_Sales',
       'EducationField_Life Sciences', 'EducationField_Marketing',
       'EducationField_Medical', 'EducationField_Other',
       'EducationField_Technical Degree', 'Gender_Male',
       'JobRole_Human Resources', 'JobRole_Laboratory Technician',
       'JobRole_Manager', 'JobRole_Manufacturing Director',
       'JobRole_Research Director', 'JobRole_Research Scientist',
       'JobRole_Sales Executive', 'JobRole_Sales Representative',
       'MaritalStatus_Married', 'MaritalStatus_Single', 'OverTime_Yes']

for col in unique_values:
    print(f"unique values in {col} are : {df[col].unique()}")
```

#### Split the columns into input and target variables (X and y)

```python
X = newdf.drop(columns=['Attrition'])
y = newdf['Attrition']
```

#### Carry out Feature scaling using StandardScaler

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
xcolumns = X.columns
X = scaler.fit_transform(X)
scaled_X = pd.DataFrame(X,columns=xcolumns)
```

```python
scaled_X.head()
```

### Split the dataset into training and testing set

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.3,random_state=100)
```

### Random Forest Classifier
** Choose the best estimator and parameters :GridSearchCV**

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import GridSearchCV
```

```python
RandomForestClassifier()
```

```python
param_grid={'max_depth': [10],
                         'n_estimators': [100], 'n_jobs': [1],
                         'random_state': [10]}
```

```python
grid_search = GridSearchCV(
    estimator=RandomForestClassifier(),
    param_grid=param_grid,
    scoring='accuracy'
)
```

```python
print(grid_search)
```

```python
grid_search = GridSearchCV(RandomForestClassifier(),param_grid,cv=5)
grid_search.fit(X_train, y_train)
```

```python
best_estimator = grid_search.best_estimator_
best_params = grid_search.best_params_
best_score = grid_search.best_score_


print(f"Best estimator is:  {best_estimator}")
print(f"Best parameters are:  {best_params}")
print(f"Best score is:  {best_score:.5f}")
```

<span style = "color:blue"> Create Random forest model with the best parameters </span>

```python
model = RandomForestClassifier(max_depth=10, n_estimators=32, n_jobs=1, random_state=20)
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
accuracy = accuracy_score(y_test, y_pred)
print(f"Accuracy Score: {best_score:.5f}")

conf_matrix = confusion_matrix(y_pred, y_test)
print("\nConfusion Matrix:")
print(conf_matrix)

class_report = classification_report(y_test, y_pred)
print("\nClassification Report:")
print(class_report)


cv_scores = cross_val_score(best_estimator, X, y, cv=5, scoring='accuracy')

print(f"Cross val score.: {cv_scores.mean():.5f}")
```
